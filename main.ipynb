```python
def frequencytable(text, k):
    freqmap = {}
    for i in range(len(text)-k):
        pattern = text[i:i + k]
        if pattern not in freqmap:
            freqmap[pattern] = 1
        else:
            freqmap[pattern] += 1
    return freqmap
```

```python
text = "ACGTTGCATGTCGCATGATGCATGAGAGCT"
k = 4
test_dict = frequencytable(text, k)
max_freq = max(test_dict.values())
[key for key, value in test_dict.items() if value == max_freq]
```
